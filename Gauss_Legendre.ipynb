*Gauss-Legendre Method*


```python
#@title Nodos y Pesos de Gauss-lagrange en $[-1,1]$
import numpy as np, math
import matplotlib.pyplot as plt


#Nodos y pesos
t = np.array([-math.sqrt(3/5),0.0, math.sqrt(3/5)])
w = np.array([5/9,8/9,5/9])

def gauss_G3(f,a,b):
  mid = 0.5*(a+b)
  med = 0.5*(b-a)

  x_nodo = mid + med * t #mapeo
  fx = f(x_nodo)
  aprox = med * np.dot(w,fx)
  return aprox, x_nodo, fx
```

```python
#@title Evaluación de la Integral $(b) \int_{1}^{3} \ \ln(x)dx$
f_b = lambda x: np.log(x)
I_b_aprox, nodo_b, fx_b = gauss_G3(f_b,1,3)
print(f"El valor de la integral (b) es: {I_b_aprox:0.8f}")
```

```python
#@title Gráfico
plt.figure(figsize=(8,6))


plt.subplot(3,1,2)
xx = np.linspace(1,3,400)
plt.plot(xx, f_b(xx))
plt.scatter(nodo_b, fx_b, color='red')
plt.title('(b) ln(x)')


plt.tight_layout()
plt.show()
```
