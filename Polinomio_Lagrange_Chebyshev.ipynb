```python
#@title Polinomio de Lagrange

import numpy as np
import matplotlib.pyplot as plt

#Definela base de LAgrange
def lagrange_base(x,j,x_nodo):
  "Calcula el polinomio de base de Lagrange L_j(x)"
  xj = x_nodo[j]
  L = np.ones_like(x)
  for i,xi in enumerate(x_nodo):
    if i!=j:
      L *=(x-xi)/(xj-xi)
  return L
```
