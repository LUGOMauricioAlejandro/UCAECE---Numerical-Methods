*Gauss-Legendre Method*


```python
#@title Definición de Función


#función objetivo
f = lambda x: 1/(1+x**2)

#valor exacto
I_exacto = np.pi / 2
```

```python
#@title Gauss-Legendre n=3
x3 = np.array([-np.sqrt(3/5),0,np.sqrt(3/5)]) #Raices del polinmio de grado 3
w3 = ([5/9,8/9,5/9]) #tres raíces
I3 = np.sum(w3*f(x3))
```

```python
#@title Imprime resultados
print(f"Valor exacto: I = π/2 = {I_exacto:.10f}")
print(f"Gauss-Legendre n=3: {I3:.10f}  Error = {abs(I_exacto - I3):0.2e}")
```
