*ODE Euler Method - PVI*

```python
#@title Código en Python
import numpy as np
import matplotlib.pyplot as plt

#Definición del problema
def f(t,x):
  return -2*t*x

#Solución exacta
def x_exacta(t):
  return np.exp(-t**2)

#Parámetros
h = 0.0085
t0,tf = 0,1
N = int((tf-t0)/h)
t = np.linspace(t0,tf, N+1)
```

```python
#@title Método de Euler
x = np.zeros(N+1)
x[0] = 1
for n in range(N):
  x[n+1] = x[n] + h*f(t[n],x[n])

#Solución Exacta
x_vexacto = x_exacta(t)

#Error en t=1
error = abs(x_vexacto[-1]-x[-1])
```
