*explicit Euler method  is the simplest numerical method to solve an ordinary differential equation*

```python
import numpy as np
import matplotlib.pyplot as plt

#Valores para el método de Euler
h = 0.01
t0 = 0.0
tf = 10.0
N = int(np.ceil((tf-t0)/h))
h = (tf-t0)/N
t = np.linspace(t0,tf,N+1)

#definición
def f(t,x):
  return (x-5.0)*(np.cos(t)**2 - 0.5)
```

```python
#@title Método de Euler en forma Explícita
ks = list(range(0,11)) #determianr la malla 10*10
solucion = {}
for k in ks:
  x = np.zeros(N+1)
  x[0] = k
  for n in range(N):
    x[n+1] = x[n] + h*f(t[n],x[n])
  solucion[k] = x #alamcena los resultados
```

```python
#@title Gráfico para $[0,10] \times [0,10]$

plt.xlim(0, 10)
plt.ylim(0, 10)
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Soluciones por Euler explícito para k=0,1,...,10 (h=0.01)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
